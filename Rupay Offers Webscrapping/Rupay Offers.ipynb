# Main Code

```python
DF = pd.DataFrame(columns=['Name','Description','Icon','Terms','Validity','Url'])
x=0
for urlL in all_urls:
    for url in urlL:
        all=[]
        source_code = requests.get(url).text
        soup = bs(source_code,'lxml')
        name=[]
        for a in soup.find_all('span',{'class':"inline even last"}):
            name.append(a.text.replace('\n','').replace('  ',''))
        all.append(name[0])
        desc=[]
        for b in soup.find_all('div',{'id':'tab1'}):
            desc.append(b.text)
        code = []
        for c in soup.find_all('h4'):
            code.append(c.text)
        description = str(desc[0].replace('\n','').replace('  ','')) + '\n' +str(code[0].replace('\n','').replace('  ',''))
        all.append(description)
        terms=[]
        for d in soup.find_all('div',{'id':'tab3'}):
            terms.append(d.text)
        date=[]
        for e in soup.find_all('div',{'id':'tab2'}):
            date.append(e.text)
        termsc = str(terms[0])
        img=[]
        for f in soup.find_all('img',{'class':"img-responsive"}):
            img.append(f['src'])
        all.append(img[1])
        all.append(termsc)
        all.append('Offer Validity :'+ str(date[0].replace('\n','').replace('  ','')))
        all.append(url)
        DF.loc[x] = all
        x = x+1
```

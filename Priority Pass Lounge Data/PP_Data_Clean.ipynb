```
def unique(list1): 
  
    # intilize a null list 
    unique_list = [] 
      
    # traverse for all elements 
    for x in list1: 
        # check if exists in unique_list or not 
        if x not in unique_list: 
            unique_list.append(x) 
    # print list 
    return unique_list
```

```
l = unique(l)
l
```

```
for num in l:
  pp_data[num] = pp_data.apply(lambda _: np.nan, axis=1)
```

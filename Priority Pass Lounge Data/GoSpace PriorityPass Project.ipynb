```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs
import requests

import time

source_code = requests.get('https://www.prioritypass.com/en/airport-lounges').text
soup = bs(source_code,'lxml')
l=[]
url_list=[]
for a in soup.find_all('div',{'class':'padded-sides'}):   
        for b in a.find_all('a',href=True):
            try:
                url_list.append("https://www.prioritypass.com/"+str(b["href"]))  
            except AttributeError:
                continue

url_list = [x for x in url_list[25:] if x != 'https://www.prioritypass.com/#' ]
url_list = url_list[:-20]
```

```python
url_list
```

```python
url_listoflists=[]
i=0
for u in url_list:
        urllist = []
        source_code = requests.get(u).text
        soup = bs(source_code,'lxml')
        for a in soup.find_all('div',{'class':"content"}):
          if(type(a) is not NoneType):
            for c in a.find('ul').find_all('a',href = True):
              if(type(c) is not None):
                try:
                    urllist.append("https://www.prioritypass.com/"+str(c["href"]))
                except AttributeError:
                    continue
        urllist = urllist[:-10]
        urllist =urllist[:round(len(urllist)/2)]
        url_listoflists.append(urllist)  
        print('{} ( {} ) '.format(len(url_listoflists),len(url_listoflists[i])))
        i=i+1
url_listoflists
```

```python
df_url = pd.DataFrame.from_records(url_listoflists)
df_url.to_csv('/content/drive/My Drive/Colab Notebooks/Lounge_Air/mynt/url_file.csv')
```

```python
x=1
DF = pd.DataFrame(columns=['Name','Airport','Country','Facilities','MainUrl','ImageUrl','Opening Hour','Location','Conditions','Additional Info'])
for urlL in url_listoflists:
    for url in urlL:
        source_code = requests.get(url).text
        soup = bs(source_code,'lxml')
        all=[]
        try:
            for a in soup.find_all('div',{'id':"lounge-header"}):
                name=a.text[1:]
                for i in (0,1):
                    all.append(name.split('\n')[i])
            all.append((url.split('/')[-3]).upper())
            facilities_list=[]
            facilities=[]
            for a in soup.find_all('figure',{'class':"img-holder"}):
                facilities_list.append(a.text)
            for i in range(0,len(facilities_list)):
                    facilities.append(facilities_list[i].replace("\n", ""))
            all.append(facilities)
            all.append(url)
            img_url=[]
            for a in soup.find_all('img',{'class':'lounge-image'},src=True):
                img_url.append(a['src'])
            all.append(img_url)
            for a in soup.find_all('article',{'class':"detail-box"}):
                part_info = a.find('span')
                all.append(part_info.text.replace('\r','').replace('\n',''))
            if len(all)==9:
                all.append('None')
                DF.loc[x] = all
                x=x+1
                print('{}. {} done'.format(x-1,url))
            else:
                DF.loc[x] = all
                x=x+1
                print('{}. {} done with Additional Info'.format(x-1,url))
        except requests.exceptions.SSLError as e:
            print(e)
            continue
```

```
File_list  = os.listdir('.')
File_list
```

```
place_data1 = pd.read_excel(File_list[0],sheet_name=0, header=0)
place_data2 = pd.read_excel(File_list[0],sheet_name=1, header=0)
place_data1.columns
```

```
dfi = pd.DataFrame(columns = place_data1.columns)
dfo = pd.DataFrame(columns = place_data2.columns)
for i in range(len(File_list)):
  place_data1 = pd.read_excel(File_list[i],sheet_name=0, header=0)
  place_data2 = pd.read_excel(File_list[i],sheet_name=1, header=0)
  dfi = dfi.append(place_data1)
  dfo = dfo.append(place_data2)
  print(File_list[i].split('.')[0], " ------------------------------------------------Apended")
dfi = dfi.reset_index(drop=True)
dfo = dfo.reset_index(drop=True)
dfi['Id'] = dfi.index
```

```
writer = pd.ExcelWriter('Asia.xlsx', engine='xlsxwriter',options={'strings_to_urls': False})

dfi.to_excel(writer,sheet_name='input',index=False)
dfo.to_excel(writer,sheet_name='Output',index=False)
writer.save()
```

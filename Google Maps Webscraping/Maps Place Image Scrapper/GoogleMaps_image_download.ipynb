```
def url_to_png(i,url,file_path):
  filename = 'image {}.png'.format(i)
  full_path = '{}{}'.format(file_path,filename)
  urllib.request.urlretrieve(url, full_path)
  print('{} saved'.format(full_path))
```

```
def image_scrapper(ul,name,index):
  options = webdriver.ChromeOptions()
  options.headless = True
  #options.add_argument('--headless')
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  prefs = {"profile.managed_default_content_settings.images": 2}
  options.add_experimental_option("prefs", prefs)
  options.add_experimental_option("excludeSwitches", ["enable-logging"])
  driver = webdriver.Chrome('chromedriver',options=options)
  driver.set_window_position(0, 0)
  driver.set_window_size(1024, 768)
  print ('Settings set!')
  print (70 * '-')
  driver.get(ul)
  print('Starting For : ',ul)
  driver.implicitly_wait(3)
  bb = driver.find_element_by_xpath("//button[@class='section-hero-header-image-hero widget-pane-fading widget-pane-fade-in section-hero-header-image-hero-clickable']")
  driver.execute_script("arguments[0].click()", bb)
  time.sleep(2)
  i = 0
  while True:
    images = driver.find_elements_by_css_selector("div[class='gallery-image-high-res loaded']")
    driver.execute_script("arguments[0].scrollIntoView();", images[-1])
    if(i == 40):
      break
    else:
      i = i+1
  image_u = []
  for i in images:
    image_u.append(i.get_attribute('style'))
  #print(len(images)) 
  image_urls = []
  for image in image_u:
    url = re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', image)
    image_urls.append(url[0])
  #print(image_urls)
  print(len(image_urls),"Url links Appended")
  file_p = "images/" + str(index) + "." + str(name) + "/"
  os.makedirs(file_p)
  for i in range(len(image_urls)):
    url_to_png(i,image_urls[i],file_p)
  print (70 * '-')
  return file_p
  driver.close()
```
